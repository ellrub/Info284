```python
from sklearn.neighbors import KNeighborsClassifier
from pandas import read_csv
from pandas import get_dummies
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
path = "./data/DATASET_REDUX.csv"
dataset = read_csv(path)
#dataset = read_csv(path,usecols=["Reviewer_Score", 'Review_Total_Negative_Word_Counts', 'Review_Total_Positive_Word_Counts'])

print(dataset.head())
#newdataset = dataset.dropna()   #Removed not a number
#dummies = get_dummies(newdataset, dtype = int)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(dataset[["Review_Total_Negative_Word_Counts", "Review_Total_Positive_Word_Counts"]].values, dataset["Reviewer_Score"].values, test_size=0.20, random_state=1)
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(X_train, Y_train)
y_pred_train = knn.predict(X_train)
y_pred_test = knn.predict(X_test)
```

```python
import sklearn


matrix = sklearn.metrics.confusion_matrix(y_pred_test, Y_test)

train_accuracy = accuracy_score(Y_train, y_pred_train)
test_accuracy = accuracy_score(Y_test, y_pred_test)
overfitting = train_accuracy - test_accuracy
```

```python
print("k = 3")
print(matrix)
print(f'train_accuracy: {train_accuracy}')
print(f'test_accuract: {test_accuracy}')
print(f'overfitting: {overfitting}')
print("--------------------------------------------------")
```
